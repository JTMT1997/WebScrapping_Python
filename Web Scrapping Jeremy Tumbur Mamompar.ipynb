```python
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import os
pd.options.display.max_columns = 300
```

```python
url_hotel = "https://www.oyorooms.com/id/hotels-in-kota-tua-jakarta/"
#driver  = webdriver.Firefox()
option = webdriver.ChromeOptions()
option.add_argument("--start-maximized")

#uncomment untuk play di google colab
option.add_argument('--headless')
option.add_argument('--disable-gpu')
option.add_argument('--no-sandbox')
option.add_argument('--remote-debugging-port-9230')
#Comment 4 code diatas(option.add), jika di run di locak

driver_hotel = webdriver.Chrome(options=option)
driver_hotel.get(url_hotel)

#membuat objek BeautifulSoup dari konten html
soup = BeautifulSoup(driver_hotel.page_source, 'html.parser')
```

```python
for tag in soup.find_all(True):
  print(tag.name)
```

```python
div_containers = soup.findAll('div', attrs={'class':"oyo-cell--12-col oyo-cell--8-col-tablet oyo-cell--4-col-phone"})
print(div_containers)
```

```python
name_hotel = []
for div in div_containers:
  name = div.find('h3', attrs={'class':"listingHotelDescription__hotelName"}).text
  name_hotel.append(name)
name_hotel = pd.DataFrame(name_hotel, columns=["Name"])
name_hotel
```

```python
price_hotel = []
for div in div_containers:
  price = div.find('span', attrs={'class':"listingPrice__finalPrice"}).text
  price_hotel.append(price)
price_hotel = pd.DataFrame(price_hotel, columns=["Price"])
price_hotel
```

```python
ratings_hotel = []
for container in div_containers:
  rating = container.find('div', attrs={'class':"hotelRating"}).text.strip()
  ratings_hotel.append(rating)
ratings_hotel = pd.DataFrame(ratings_hotel, columns=["Ratings"])
ratings_hotel
```

```python
df_hotel = pd.concat([name_hotel, price_hotel, ratings_hotel], axis=1)

print('Total Hotels : ', len(df_hotel))

df_hotel
```

```python
#convert df_hotel to CSV
df_hotel.to_csv('hotel_data.csv', index=False)
#After this We Can Export to SQL Database
```
